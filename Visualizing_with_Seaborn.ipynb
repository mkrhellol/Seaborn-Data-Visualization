```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

**Loading the datasets**

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planet = sns.load_dataset("planets")
```

**Checking the Tips dataset**

```python
tips
```

**Scatter Plot**

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", size="size")
```

**Histograms & Distribution Plot**

```python
sns.histplot(tips['tip'], kde=True, bins = 20)
plt.xlabel("Tip in $")
```

**Playing with bins size**

```python
sns.histplot(tips['tip'], kde=True, bins = 10)
plt.xlabel("Tip in $")
```

```python
sns.histplot(tips['tip'], bins = 20)
plt.xlabel("Tip in $")
```

```python
sns.distplot(tips['tip'], kde=True, bins = 20)
plt.xlabel("Tip in $")
```

**Bar Plot**

```python
sns.barplot(x="sex", y= "tip", data = tips, palette ="YlGnBu")
```

**Box Plot**

```python
sns.boxplot(x="day", y="tip", data=tips, hue = "sex", palette = "YlGnBu")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,hue = "sex", palette = "YlGnBu")
```

**Strip Plot**

```python
sns.stripplot(x="day", y="tip", data=tips, palette = "YlGnBu", hue="sex", dodge = True)
```

**Joint Plot**

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="scatter", color = "r")
```

**Joint Plot with Regression**

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg", color = "g")
```

**Joint Plot with kde without and with shading**

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", color = "b")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", shade = "True", cmap = "hls")
```

**Hex type joint plot**

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex", cmap="YlGnBu")
```

**Checking the titanic dataset**

```python
titanic.head()
```

**Pair plot on titanic dataset**

```python
sns.pairplot(titanic.select_dtypes(["number"]), hue ="pclass")
```

**Preprocessing with label encoder**

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
columns = ["adult_male", "sex","alone"]
for col in columns:
    titanic[col] = le.fit_transform(titanic[col])
titanic.head()
```

**Considering only numerical values**

```python
titanicnum = titanic.select_dtypes(["number"])
titanicnum.head()
```

**Correlation of different factors**

```python
titanicnum.corr()
```

**Heatmap on titanic dataset**

```python
sns.heatmap(titanicnum.corr(), annot = True, cmap = "coolwarm")
```
